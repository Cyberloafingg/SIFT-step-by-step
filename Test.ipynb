```python
import cv2 as cv
import matplotlib.pyplot as plt
def detect_sift(img):
    sift = cv.SIFT_create() # SIFT特征提取对象
    kp = sift.detect(img, None) # 关键点位置
    kp, des = sift.compute(img, kp) # des为特征向量
    print(f"特征点个数：{len(kp)}")
    return kp, des
```

```python
image = cv2.imread('box_in_scene.png', 0)
kp1, des1 = detect_sift(image)
img = cv.drawKeypoints(image, kp1, image, color=(201,255,125), flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img)
```

```python
plt.imsave('sift_keypoints.jpg', img)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 定义正方形图像的边长
image_size = 20

# 定义图像中每个小方块的边长
block_size = image_size // 4

# 创建一个图像数组
image = np.zeros((image_size, image_size))

# 遍历每个小方块，并在其中绘制8个方向
for i in range(4):
    for j in range(4):
        x_start = i * block_size
        y_start = j * block_size
        for k in range(8):
            x_end = x_start + int(block_size/2 * np.sin(k*np.pi/4))
            y_end = y_start + int(block_size/2 * np.cos(k*np.pi/4))
            plt.plot([x_start, x_end], [y_start, y_end], linewidth=2)

# 显示图像
plt.axis('off')
plt.show()
```
